```python
import tkinter as tk
from tkinter import ttk, messagebox
from ttkthemes import ThemedTk
import pyttsx3
import speech_recognition as sr
import os
import time
import pyautogui
import pygetwindow as gw
```

```python
#PLEASE RUN THIS AFTER INSTALLING ALL MODULES

# Function to speak out text
def speak(text):
    engine = pyttsx3.init()

    # Changing voice properties to female
    voices = engine.getProperty('voices')
    engine.setProperty('voice', voices[1].id)  # Index 1 usually represents a female voice

    engine.say(text)
    engine.runAndWait()

# Function to listen for voice command
def listen():
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        print("Listening...")
        audio = recognizer.listen(source)
    try:
        command = recognizer.recognize_google(audio)
        print(f"You said: {command}")
        return command.lower()
    except sr.UnknownValueError:
        speak("Sorry, I did not understand that.")
        return ""
    except sr.RequestError:
        speak("Sorry, my speech service is down.")
        return ""

# Function to play video
def play_video(video_path):
    os.startfile(video_path)
    time.sleep(2)
    
    windows = gw.getWindowsWithTitle(os.path.basename(video_path))
    if windows:
        window = windows[0]
        window.activate()
        pyautogui.press('f')

# Function to start listening and handle commands
def start_listening():
    speak("Hello, welcome! How can I assist you today?")
    while True:
        command = listen()
        if "play the presentation" in command:
            speak("Playing the presentation.")
            play_video(r"C:\Users\HP\Desktop\BMS\inauguration\sample.pptx")
            break
        elif "exit" in command or "quit" in command:
            speak("Goodbye!")
            break

# Function to create and run the GUI
def create_gui():
    root = ThemedTk(theme="breeze")
    root.title("Voice Assistant")
    root.attributes("-fullscreen", True)

    style = ttk.Style()
    style.configure('TButton', font=('Helvetica', 12), padding=10)
    style.configure('TLabel', font=('Helvetica', 14))

    container = ttk.Frame(root)
    container.pack(expand=True)

    label_title = ttk.Label(container, text="Voice Assistant", font=("Helvetica", 20))
    label_title.pack(pady=20)

    listen_button = ttk.Button(container, text="Start Listening", command=start_listening)
    listen_button.pack(pady=10)

    exit_button = ttk.Button(container, text="Exit", command=root.destroy)
    exit_button.pack(pady=10)

    root.mainloop()

if __name__ == "__main__":
    create_gui()
```
